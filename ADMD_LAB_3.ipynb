```python
# Importing necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# Load dataset
# Original URL: url = 'https://raw.githubusercontent.com/openai/data/master/titanic.csv'
url = 'https://storage.googleapis.com/tf-datasets/titanic/train.csv'  # Changed URL to a known working source
data = pd.read_csv(url)
# 1. Generate Statistical Summary
print("Statistical Summary:")
print(data.describe())
# Summary of categorical columns
print("\nCategorical Summary:")
print(data.describe(include=['object']))
# 2. Data Visualizations using Matplotlib
# Histogram for 'Age'
plt.hist(data['age'].dropna(), bins=20, edgecolor='black', color='skyblue')  # Changed 'Age' to 'age'
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
# Bar chart for 'Pclass'
plt.figure(figsize=(6, 4))
# Changed 'Pclass' to 'pclass' and used value_counts().sort_index() for consistent ordering
# Check if the column name is 'pclass' or 'Pclass' or 'class'
pclass_column = 'pclass' if 'pclass' in data.columns else 'Pclass' if 'Pclass' in data.columns else 'class'
data[pclass_column].value_counts().sort_index().plot(kind='bar', color='lightcoral')
plt.title('Passenger Class Distribution')
plt.xlabel('Class')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
# Box plot for 'Age'
plt.figure(figsize=(6, 4))
plt.boxplot(data['age'].dropna(), patch_artist=True,  # Changed 'Age' to 'age'
            boxprops=dict(facecolor='lightgreen'))
plt.title('Box Plot of Age')
plt.ylabel('Age')
plt.show()
# 3. Data Visualizations using Seaborn
# Pair plot to visualize relationships between numerical features
# Original: sns.pairplot(data[['Age', 'Fare', 'SibSp', 'Parch']].dropna())
# Changed column names to lowercase and corrected 'sibsp' to 'n_siblings_spouses'
sns.pairplot(data[['age', 'fare', 'n_siblings_spouses', 'parch']].dropna())
plt.show()
# Correlation heatmap
# Corrected 'sibsp' to 'n_siblings_spouses' in correlation matrix calculation
correlation_matrix = data[['age', 'fare', 'n_siblings_spouses', 'parch']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', cbar=True)
plt.title('Correlation Heatmap')
plt.show()
# Count plot for 'Survived'
# Assuming the actual column name is 'survived', replace 'Survived' with 'survived'
sns.countplot(x='survived', data=data, palette='Set2')
plt.title('Survival Count')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.show()
# Violin plot for 'Age' by 'Pclass'
# Check if the column name is 'pclass' or 'Pclass' or 'class'
pclass_column = 'pclass' if 'pclass' in data.columns else 'Pclass' if 'Pclass' in data.columns else 'class'
sns.violinplot(x=pclass_column, y='age', data=data, palette='muted')  # Changed 'Pclass' to 'pclass', 'Age' to 'age'
plt.title('Age Distribution by Passenger Class')
plt.show()
```
