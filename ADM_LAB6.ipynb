```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from mpl_toolkits.mplot3d import Axes3D

```

```python
#sample dataset (3D Data)
data = {
    'Feature1':[2,4,6,8,10],
    'Feature2':[3,5,7,9,11], # Added an element to make it the same length
    'Feature3':[4,6,8,10,12]
}
#convert  to DataFrame
df = pd.DataFrame(data)
```

```python
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)

fig = plt.figure(figsize=(12, 5))
```

```python
ax1=fig.add_subplot(121,projection='3d')
ax1.set_xlabel("Feature1")
ax1.set_ylabel("Feature2")
ax1.set_zlabel("Feature3")
ax1.set_title("Before PCA (3D)")
ax1.legend()
ax1.scatter(df['Feature1'],df['Feature2'],df['Feature3'],color='blue',label="Original Data")
```

```python
pca = PCA(n_components=2)
principal_components = pca.fit_transform(scaled_data)
pca_df = pd.DataFrame(principal_components, columns=['PC1','PC2'])
```

```python
ax2 = fig.add_subplot(122)
ax2.scatter(pca_df['PC1'],pca_df['PC2'],color='red',
label="PCA Transformed Data")
ax2.set_xlabel("Principal Component 1")
ax2.set_ylabel("Principal Component 2")
ax2.set_title("After PCA (2D)")
ax2.legend()
ax2.grid()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from mpl_toolkits.mplot3d import Axes3D

data = {
    'Feature1': [2, 4, 6, 8, 10],
    'Feature2': [1, 3, 5, 7, 9],
    'Feature3': [5, 10, 15, 20, 25]
}

df = pd.DataFrame(data)

scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)

fig = plt.figure(figsize=(12, 5))
ax1 = fig.add_subplot(121, projection='3d')
ax1.set_xlabel('Feature1')
ax1.set_ylabel('Feature2')
ax1.set_zlabel('Feature3')
ax1.set_title('Before PCA (3d)')
ax1.legend()

ax1.scatter(df['Feature1'], df['Feature2'], df['Feature3'], color='blue', label='Original Data')

pca = PCA(n_components=2)
principalComponents = pca.fit_transform(scaled_data)

pca_df = pd.DataFrame(data = principalComponents, columns = ['PC1', 'PC2'])

ax2 = fig.add_subplot(122)
ax2.scatter(pca_df['PC1'], pca_df['PC2'], color='red', label='PCA Transformed data')
ax2.set_xlabel('Principal Amount 1')
ax2.set_ylabel('Principal Amount 2')
ax2.set_title('After PCA (2d)')
ax2.legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as nmp
import matplotlib.pyplot as mplt
import pandas as pnd
```

```python
DS=pnd.read_csv('wine.csv')
X=DS.iloc[:,0:13].values
Y=DS.iloc[:,13].values
```

```python
from sklearn.model_selection import train_test_split as tts
X_train,X_test,Y_train,Y_test=tts(X,Y,test_size=0.2,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler as SS
SC=SS()
X_train=SC.fit_transform(X_train)
X_test=SC.transform(X_test)
```

```python
from sklearn.decomposition import PCA
PCa = PCA (n_components=1)
X_train=PCa.fit_transform(X_train)
X_test=PCa.transform(X_test)
explained_variance=PCa.explained_variance_ratio_
```

```python
from sklearn.linear_model import LogisticRegression as LR
classifier_1 = LR(random_state = 0)
classifier_1.fit(X_train,Y_train)
```

```python
Y_pred=classifier_1.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix as CM
c_m=CM(Y_test,Y_pred)
```
